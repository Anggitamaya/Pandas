

---



                                                         
                                                          
                                                           
                                                            TUGAS SOAL-2

Buat dataframe 'jumlah_produk' yang menunjukkan jumlah produk 1 sampai 41 beerta total harga







---



```python
results =[]
for i in range(1, 41):
  kode_produk = f"PROD-00000{i:02d}"
  total_harga = sum(data[(data['kode_produk'] == kode_produk)]['harga_berlaku_cabang'])
  results.append({'kode_produk': kode_produk, 'total_harga': total_harga})

df = pd.DataFrame(results)
print(df)

data['total'] = df['total_harga']
data.head(42)
```

```python
results = []
for i in range(1, 41):
  kode_produk = f"PROD-00000{i:02d}"
  total_produk = len(data[(data['kode_produk'] == kode_produk)])
  results.append({'kode_produk': kode_produk, 'total_produk': total_harga})

df = pd.DataFrame(results)
print(df)

data['jumlah_produk'] = df['total_produk']
data.head(42)
```
